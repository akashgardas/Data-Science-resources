```python
import matplotlib.pyplot as plt
```

```python
# squares graph
x = [i for i in range(1, 5)]
y = [i**2 for i in x]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.bar(x, y)
plt.show()
```

```python
# plt.xticks([i for i in range(1, 5)])
plt.xticks(x) # scale on x-axis
plt.xlabel('x')
plt.ylabel('y')
plt.title('x-y')
plt.bar(x, y) # default orientation
plt.show()
```

```python
plt.xticks(y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x-y')
plt.barh(x, y) # horizontal orientation
plt.show()
```

```python
plt.barh(x, y)
plt.show()
```

```python
# sine and cosine graph
import math
x = [0, 30, 45, 60, 90]
siny = [math.sin(i) for i in x]
cosy = [math.cos(i) for i in x]
plt.xticks(x)
plt.xlabel('degree')
plt.ylabel('value')
plt.plot(x, siny, label='sine')
plt.plot(x, cosy, label='cosine')
plt.legend()
plt.show()
```

```python
# markers 
# and line styles
```

```python
plt.plot(x, siny, 'rD', label='sine')
plt.plot(x, cosy, 'bD', label='cosine')
plt.show()
```

```python
plt.plot(x, siny, 'r-.', label='sine')
plt.plot(x, cosy, 'g-.', label='cosine')
plt.show()
```

# Distribution plot / Histogram

```python
import random
# collecting random sample
l1 = list(random.sample(range(1, 500), 100)) # 100 samples
plt.hist(l1, bins=20) # bins
plt.show()
```

```python
plt.hist(l1, histtype='step') # histtype
plt.show()
```

```python
plt.hist(l1, histtype='barstacked') # histtype
plt.show()
```

# Box plot

```python
plt.boxplot(l1) # min value, 25 % value, avg value, 75 % value, max value
plt.show()
```

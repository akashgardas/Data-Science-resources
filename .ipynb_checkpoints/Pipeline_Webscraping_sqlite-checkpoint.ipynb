## Fetching data from a HTML file and storing in this database

```python
# building a pipeline
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
result = requests.get('https://quotes.toscrape.com/')
result
```

```python
result.text # it's an html page. So, we use BeautifulSoup lib
```

```python
# converting into a meaningful data
data = BeautifulSoup(result.text, 'html.parser')
data
```

```python
data.find_all('a') # finds all <a></a> tags
```

```python
data.find_all('div')
```

```python
data.find_all('div', class_='quote')
```

```python
data.find_all('div', class_='quote')[0].find('span', class_='text').text
```

```python
# finding all the quotes in the page
for div in data.find_all('div', class_='quote'):
    print(div.find('span', class_='text').text)
```

```python
data.find_all('div', class_='quote')[0].find('small', class_='author').text
```

```python
# fetching the author names
for div in data.find_all('div', class_='quote'):
    print(div.find('small', class_='author').text)
```

```python
# storing the author name and quote into a dict
data1 = data.find_all('div', class_='quote')
data = {}
for quote in data1:
    quotation = quote.find('span', class_='text').text
    authorname = quote.find('small', class_='author').text
    data[authorname] = quotation
data
```

```python
# creating a sqlite table 
cursor.execute('create table qtable(authorname text, quotation text)')
```

```python
# inserting the data into the table
for x in data:
    cursor.execute('''
    insert into qtable(authorname, quotation)
    values(?, ?)
    ''', (x, data[x]))
```

```python
records = cursor.execute('select * from qtable')
records
```

```python
for row in records:
    print(row)
```
